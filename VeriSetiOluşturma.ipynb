```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('/content/user_behavior_dataset.csv')
İşletimSistemi= veriSeti['Operating System']
Cihaz = veriSeti['Device Model']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(İşletimSistemi , Cihaz)
plt.xlabel('İşletim Sistemi')
plt.ylabel('Cihaz Modeli')
plt.title('Cihazların İşletim Sistemlerini Kullanım Oranları')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```
